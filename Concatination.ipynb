```python
class FileLoad:
    csv1_path='/content/drive/MyDrive/MachineLearningEngineerTest/Java/resources/1.csv'
    csv2_path='/content/drive/MyDrive/MachineLearningEngineerTest/Java/resources/2.csv'
    output_path='/content/drive/MyDrive/MachineLearningEngineerTest/Java/resources/output.csv'

    def combineFiles(self,csv1,csv2,outputFile):
        #reading csv files in pandas dataframe
        df1=pd.read_csv(csv1)
        df2=pd.read_csv(csv2)

        # merging both csv file using variables in 6414 column
        merged_df = df1.merge(df2, on='6414', how='inner')

        #converting merged dataframe in output.csv
        merged_df.to_csv(outputFile, index=False)

        print("Both of the csv are merged sucessfully")

```

```python
if __name__ == '__main__':
  csv1=FileLoad.csv1_path
  csv2=FileLoad.csv2_path
  output_csv=FileLoad.output_path
  merge=FileLoad()
  merge.combineFiles(csv1,csv2,output_csv)

```
